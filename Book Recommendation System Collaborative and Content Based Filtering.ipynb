Importing the Required Packages 

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Reading the datas 

```python
books=pd.read_csv("books.csv")
```

```python
book_tags=pd.read_csv("book_tags.csv")
```

```python
tags=pd.read_csv("tags.csv")
```

Attaching tag names to each row of the book data 

```python
book_tags = book_tags.merge(tags, on="tag_id", how="left")
```

```python
print(book_tags) # to have a look 
```

```python
top_tags=(book_tags.sort_values(['goodreads_book_id','count'],ascending=[True,False]).groupby('goodreads_book_id').head(5))
```

```python
print(top_tags)
```

Explaining the merge function:
v r merging top_tags with books dataset on the basis of the goodreads_book_id column on the Right( left shld be written).

```python
books_tags_joined=books.merge(top_tags,on="goodreads_book_id",how="left")
```

```python
print(books_tags_joined)
```

COMBINING ALL TAGS OF THE BOOK INTO A SINGLE STRING

books_tags_joined.groupby(['book_id','title','authors'])
Groups all rows by book_id, title, and authors.
So for each book, you collect all its tags together.
----------
['tag_name']
From each group, we only care about the tag_name column (the actual text labels of tags like “fantasy”, “romance”, etc.).
-----------------
.apply(lambda x: " ".join(x.dropna()))
For each group (one book), take the column of tag names (x is a Series of tag names).

x.dropna() removes any missing values.

" ".join(...) 
combines all the tags into a single string separated by spaces.
Example: ["fantasy", "magic", "adventure"] → "fantasy magic adventure"
This way, each book ends up with one “document” of its tags.
-----------
.reset_index()

After grouping, the grouping keys (book_id, title, authors) become the index.

reset_index() turns them back into normal columns.

```python
books_with_tags=books_tags_joined.groupby(["book_id","title","authors"])["tag_name"].apply(lambda x: " ".join(x.dropna())).reset_index()
```

```python
print(books_with_tags)
```

Building A single text field for each book ie combining the title authors and the tagnames inside a new column named features

```python
books_with_tags["features"]=(books_with_tags["title"]+" "+books_with_tags["authors"]+" "+books_with_tags["tag_name"])
```

```python
print(books_with_tags["features"])
```

```python
tfidf=TfidfVectorizer(stop_words="english")
```

```python
tfidf_matrix=tfidf.fit_transform(books_with_tags['features'])
```

```python
cosine_sim=cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
indices=pd.Series(books_with_tags.index,books_with_tags["title"]).drop_duplicates()
```

```python
print(indices)
```

Writing the Recommendation Function 

```python
def recommend_books(title,n=5):
    if title not in indices:
        return"Book not found in Dataset"
    idx=indices[title]
    sim_scores=list(enumerate(cosine_sim[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
    sim_scores=sim_scores[1:n+1]
    book_indices=[i[0] for i in sim_scores]
    return books_with_tags[["title","authors"]].iloc[book_indices]
```

```python
recommend_books("The Fault in Our Stars",5)
```

*COLLABORATIVE FILTERING*

importing required libraries for collaborative filtering

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

to prevent memory crashes we filter out the users who haven't rated many books and books that haven't been rated by many people. 

```python
#to load the ratings data
ratings=pd.read_csv("ratings.csv")
```

```python
#filtering out books fewer than 50 ratings
book_rating_counts=ratings['book_id'].value_counts()
popular_books=book_rating_counts[book_rating_counts>=50].index
ratings_filtered=ratings[ratings['book_id'].isin(popular_books)]
```

```python
#filtering users who have rated less than 50 books 
user_rating_counts=ratings['user_id'].value_counts()
active_users=user_rating_counts[user_rating_counts>=50].index
ratings_filtered=ratings_filtered[ratings_filtered['user_id'].isin(active_users)]
```

```python
#merge books dataframe to get the titles
ratings_with_titles=ratings_filtered.merge(books[['book_id','title']],on='book_id')
```

Creating the item-user sparse matrix: 
BOOKS are rows 
USERS are columns

```python
#pivot the table: Books as rows, users as columns
book_user_matrix=ratings_with_titles.pivot_table(index='title',columns='user_id',values='rating').fillna(0)
```

```python
#converting to A Scipy Sparse matrix so thet the millions of empty zeroes dont take up more space
book_user_sparse=csr_matrix(book_user_matrix.values)
```

```python
#Save the index so we can look up rows by book title later
book_title_indices=list(book_user_matrix.index)
```

Training the NEAREST NEIGHBORS MODEL

```python
model_knn=NearestNeighbors(metric="cosine",algorithm='brute')
model_knn.fit(book_user_sparse)
```

Writing the Collaborative Recommendation Function

```python
def recommend_collaborative(book_title, n=5):
    if book_title not in book_title_indices:
        return "Book not found in the filtered dataset."

    book_idx = book_title_indices.index(book_title)

    # calculating the nearest neighbours
    distances, indices = model_knn.kneighbors(
        book_user_matrix.iloc[book_idx, :].values.reshape(1, -1),
        n_neighbors=n+1
    )

    recommended_books = []   # list to store book titles

    print(f"Users who liked '{book_title}' also liked:\n")

    for i in range(1, len(distances.flatten())):
        neighbor_idx = indices.flatten()[i]
        neighbor_title = book_title_indices[neighbor_idx]
        distance_score = distances.flatten()[i]

        # similarity percentage
        similarity = (1 - distance_score) * 100

        recommended_books.append(neighbor_title)   # add book to list

        print(f"{i}: {neighbor_title} (Similarity: {similarity:.2f}%)")


    return recommended_books
    
```

```python
recommend_collaborative("The Fault in Our Stars", 5)
```
